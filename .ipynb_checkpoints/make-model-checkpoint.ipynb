## Exploratory Data Analysis

### Fetching and understanding the data

```python
# file path
path = r'X:\MSA_Phase_3_Submission\data_science_submission_phase3\data\cifar-10-python\cifar-10-batches-py\data_batch_1'

# Assume tar.gz file to be extracted at project directory.
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

# Get data
batch1_data = unpickle(path)

# Tells us data is a dictionary
#print(type(batch1_data)) 

# Get keys
print(batch1_data.keys())

# Get data shape
print ('Current batch label ðŸ‘‡')
print(batch1_data[b'batch_label'])

print ('Length of labels listðŸ‘‡')
print(len(batch1_data[b'labels']))

print ('data shape ðŸ‘‡')
print(batch1_data[b'data'].shape)

print ('Length of filenames list ðŸ‘‡')
print(len(batch1_data[b'filenames']))
```

### Reshape data

```python
# Reshape image data together
data = batch1_data[b'data']
data = data.reshape(len(data), 3, 32, 32).transpose(0, 2, 3, 1)

# Reshape label data
label_data = np.array(batch1_data[b'labels'])
label_data = label_data.reshape(len(label_data))

# Verify shape
print(data.shape)
print(label_data.shape)

#plt.imshow(data[500])
```

### Visualise images from each label

```python
# lists to contain image samples
airplane_samples = []
automobile_samples = []
bird_samples = []
cat_samples = []
deer_samples = []
dog_samples = []
frog_samples = []
horse_samples = []
ship_samples = []
truck_samples = []

# tracker to stop while loop
sample_sum = 0

# tracker to keep track of label positions in data
label_pos = 0

while sample_sum != 50:
    # Condition labels and append to appropriate sample lists
    if batch1_data[b'labels'][label_pos] == 0 and len(airplane_samples) < 5:
        airplane_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 1 and len(automobile_samples) < 5:
        automobile_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 2 and len(bird_samples) < 5:
        bird_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 3 and len(cat_samples) < 5:
        cat_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 4 and len(deer_samples) < 5:
        deer_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 5 and len(dog_samples) < 5:
        dog_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 6 and len(frog_samples) < 5:
        frog_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 7 and len(horse_samples) < 5:
        horse_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 8 and len(ship_samples) < 5:
        ship_samples.append(data[label_pos])
    elif batch1_data[b'labels'][label_pos] == 9 and len(truck_samples) < 5:
        truck_samples.append(data[label_pos])
    
    # Increment both trackers
    sample_sum = sum([len(airplane_samples),
                      len(automobile_samples),
                      len(bird_samples),
                      len(cat_samples),
                      len(deer_samples),
                      len(dog_samples),
                      len(frog_samples),
                      len(horse_samples),
                      len(ship_samples),
                      len(truck_samples)])
    label_pos +=1
    #print(sample_sum)


#print(batch1_data[b'labels'])
```

```python
# Increase plot figure size
plt.figure(figsize=(15, 35), dpi=80)

# Add arrays to plot together
plot_array = airplane_samples + automobile_samples + bird_samples + cat_samples + deer_samples + dog_samples + frog_samples + horse_samples + ship_samples + truck_samples
# Lable list
classes = ['aiplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

# label tracker
lab_track = 0

# Plot all images
for x in range(50):
    # Render image on specified subplot
    plt.subplot(10, 5, x+1)
    plt.imshow(plot_array[x])
    # Display label
    plt.xlabel(classes[lab_track])
    if ((x + 1)%5) == 0:
        lab_track += 1
```

### Reformat some data

```python
train_y, test_y = np.array(train_y), np.array(test_y)
type(train_y)
```

#### Data Modelling

```python
import tensorflow as tf
from tensorflow.keras import Input, layers, Sequential, optimizers, losses, callbacks
from datetime import datetime
```

```python
def generateModel():
    """
    Make a Keras Sequential Model.
    Feel free to change the layout of the layers.

    Read more about keras here: https://keras.io/api/
    Tensorflow setup adapted here: https://www.tensorflow.org/tutorials/quickstart/beginner
    """
    '''
    model = Sequential(name="Image_Recognition_Model")
    model.add(Input(shape=(32,32,3))) # Input layer.

    model.add(layers.Rescaling(scale=1./255, name="Normaliser")) # Example pre-processing layer.

    model.add(layers.Dense(40, activation="relu")) # Example Hidden Layer 1. Has 40 Nodes.
    model.add(layers.Dense(25, activation="sigmoid")) # Example Hidden Layer 2. Has 25 Nodes.
    model.add(layers.Dense(15, activation="tanh")) # Example Hidden Layer 3. Has 15 Nodes.

    model.add(layers.Dense(1, activation="sigmoid", name="Output_Probability")) # Output layer.
    '''

    model = Sequential(name="Image_Recognition_Model")
    model.add(layers.Flatten(input_shape=(32,32,3))) # input layer modified to accept data shape of (32, 32, 3)
    model.add(layers.Rescaling(scale=1./255, name="Normaliser")) # Example pre-processing layer.
    model.add(layers.Dense(1000, activation='relu')) # hidden layer 1
    model.add(layers.Dense(1000, activation='relu')) # hidden layer 2
    model.add(layers.Dense(800, activation='relu')) # hidden layer 3
    model.add(layers.Dense(10, activation='sigmoid', name="Output_Probability")) # output layer
    
    model.compile(
        optimizer=optimizers.SGD(learning_rate=0.01),
        loss=losses.MeanAbsoluteError(),
        metrics=['accuracy']
    )

    return model
```

```python
model = generateModel()

log_dir = "output/logs/test_8_40am_9Sep" #+ datetime.now().strftime("%Y-%m-%d-%H:%M:%S")

# Callbacks for QoL.
tensorboard_callback = callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
earlystop_callback = callbacks.EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True)

model.fit(train_x, train_y,
    validation_data=(test_x, test_y),
    epochs= 50,
    batch_size=1000, # Set this to a lower number if your computer spec is bad.
    callbacks=[tensorboard_callback, earlystop_callback],
    #verbose=0 # Uncomment this line to reduce the printing on console.
)
```
