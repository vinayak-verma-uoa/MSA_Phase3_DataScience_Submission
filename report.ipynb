### Loss

For my loss function I picked BinaryCrossentropy. My Goal is to detect whether there is a Cat in an image which is a Binary Classification problem. Therefore using BinaryCrossentropy is a better approach. Before I was using SparseCategoricalCrossentropy and this caused my loss to remain constant throghout the model fitting process. This happened because SparseCategoricalCrossentropy is suited for solving problems with multi classes.

### Optimizer

I choose the Adam optimizer as it is commonly used across data science tutorials and is also an adaptive optimizer. This makes the Adam optimizer computationally efficient, especially for people like me who have low PC specs. Before when I was using Stochastic Gradient Descent (SGD) and the convergence of my model was very slow. This can be countered by using adaptive optimizers where the learning rates change dynamically to become small only as you come closer to convergence. Therefore, using Adam saved me time by reducing model training time.
